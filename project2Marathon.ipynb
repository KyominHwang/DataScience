```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data = pd.read_csv("marathon/marathon_2015_2017.csv")
data.head()
```

```python
data.columns
```

```python
data = data[["M/F","Age","Pace","10K","20K","30K","Official Time"]]
```

```python
data["M/F"] = data["M/F"].replace("M",1).replace("F",0)
```

```python
data.head()
```

```python
data.info()
```

```python
data_list = data.values.tolist()
data_list
```

```python
x_train = [r[0:3] for r in data_list]
y_train = [r[-1] for r in data_list]
```

```python
type(x_train[0])
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1, input_shape = (3,)))
```

```python
sgd = tf.keras.optimizers.SGD(lr = 0.000001)
```

```python
model.compile(loss = "mean_squared_error", optimizer = sgd)
```

```python
model.fit(np.array(x_train), np.array(y_train), epochs = 20)
```

```python
model.predict([[1,25,296]])
```
