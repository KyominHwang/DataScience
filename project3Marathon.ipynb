```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
data = pd.read_csv("marathon/marathon_2015_2017.csv")
```

```python
data["M/F"] = data["M/F"].replace("M",1).replace("F",0)
data.head()
```

```python
data_train = data[data["Year"] != 2017]
data_test = data[data["Year"] == 2017]
```

```python
data_train = data_train[["M/F","Age","Pace","10K","20K","30K","Official Time"]]
data_test = data_test[["M/F","Age","Pace","10K","20K","30K","Official Time"]]
```

```python
data_train_list = data_train.values.tolist()
data_test_list = data_test.values.tolist()
```

```python
x_train = [r[0:3] for r in data_train_list]
y_train = [r[3:7] for r in data_train_list]

x_test = [r[0:3] for r in data_test_list]
y_test = [r[3:7] for r in data_test_list]
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(4, input_shape=(3,)))
sgd = tf.keras.optimizers.SGD(lr = 0.00001)
model.compile(loss = "mse", optimizer = sgd)

model.fit(x_train, y_train, epochs = 100)
```

```python
model.predict(x_test)
```
